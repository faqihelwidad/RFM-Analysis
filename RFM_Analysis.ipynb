```python
port pandas as pd

df = pd.read_csv('ecommerce_sales_34500.csv')

df.head()
```

```python
df.info()
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
```

```python
reference_date = df['order_date'].max() + pd.Timedelta(days=1)
```

```python
df['recency'] = (reference_date - df.groupby('customer_id')['order_date'].transform('max')).dt.days
```

```python
df.head()
```

```python
df['frequency'] = df.groupby('customer_id')['order_id'].transform('count')
```

```python
df.head()
```

```python
df['monetary'] = df.groupby('customer_id')['total_amount'].transform('sum')
```

```python
df.head()
```

```python
df['R'] = pd.qcut(df['recency'], 5, labels=[5,4,3,2,1])
df['F'] = pd.qcut(df['frequency'], 5, labels=[1,2,3,4,5])
df['M'] = pd.qcut(df['monetary'], 5, labels=[1,2,3,4,5])
```

```python
df.head()
```

```python
def assign_segment(row):
    R, F, M = row['R'], row['F'], row['M']

    if R >= 4 and F >= 4 and M >= 4:
        return "Champions"
    elif R >= 3 and F >= 4:
        return "Loyal Customers"
    elif R >= 4 and F <= 3:
        return "Potential Loyalists"
    elif R == 5 and F <= 2:
        return "New Customers"
    elif R == 4 and F <= 2:
        return "Promising"
    elif R == 3 and F == 3:
        return "Need Attention"
    elif R <= 2 and F >= 3 and M >= 4:
        return "Can't Lose Them"
    elif R <= 2 and F >= 3:
        return "At Risk"
    elif R == 2 and F <= 2:
        return "About To Sleep"
    elif R == 1 and F <= 2:
        return "Lost Customers"
    else:
        return "Others"
```

```python
df['segment'] = df.apply(assign_segment, axis=1)
```

```python
df.head()
```

```python
df['segment'].value_counts()
```

```python
segment_profile = df.groupby('segment').agg({
    'recency' : 'mean',
    'frequency' : 'mean',
    'monetary' : 'mean'
}).round(2)

segment_profile
```

```python
df.groupby('segment').agg({
    'customer_id':'nunique',
    'monetary':'sum'
}).sort_values('monetary', ascending=False)
```

```python
revenue = df.groupby('segment')['monetary'].sum()
pct = revenue / revenue.sum() * 100

pd.concat([revenue, pct], axis=1)
```
